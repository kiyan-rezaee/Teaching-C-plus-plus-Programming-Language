```python
#include <iostream>
#include <iomanip>

using namespace std;



int main()
{
    int i;
    for (i=0; i <= 10; i++)
    {
        if (i > 5)  break;
        cout << "\n" << i;
    }

    int j;
    for (j=0; j <= 10; j++)
    {
        if (j > 5)  continue;
        cout << "\n" << j;
    }

    cout << "\n" << i << " " << j << endl;

    return 0;
}
```
