```python
#include <iostream>
using namespace std;

// test(5) = test(4) + test(3) => test(3) -> 2 + 1 , test(4) = 3 + 2


int test(int n)
{
    if (n < 2) return 1;
    else if (n == 2) return 2;
    else return (test(n-1) + test(n-2));
}
int main(){
cout << test(5) << endl;
return 0;
}
```

```python
#include <iostream>
using namespace std;

// test(5) = test(4) + test(3) => test(3) -> 2 + 1 , test(4) = 3 + 2

int sqp (int x)
{
    int c=1, s=0, n=0;
    while(s<x)
    {
        s += c;
        n++;
        c += 2;
        cout << n << endl;
    }
    return(n);
}

int main()
{
    cout << sqp(5);
    return 0;
}
```

```python
#include <iostream>
using namespace std;

// 15 -> 1 + countt(1) -> 1 + 1 + countt(0) -> 1+1+0=2

int countt (int n){
    if (n == 0)
        return 0;
    else
        return 1 + countt(n/10);
}

int main()
{
    int x;
    if (x > 0)
        cout << countt(x) <<endl;
    else
        cout << 1;
    return 0;
}
```
