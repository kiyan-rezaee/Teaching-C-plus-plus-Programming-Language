```python
#include <iostream>


using namespace std;

// 132 -> 2, 20 + 3, 230 + 1 = 231
/*

sum = sum * 10 + rem => sum = 2, n / 10 = 13, 3, sum = 2 * 10 + 3 = 23
*/

int main()
{
    int n, m, rev=0;
    cin>>n;
    m=n;
    while(n>0){
          rev=n%10+ rev*10;
n=n/10;
    }
    if (m==rev)
        cout<<"not";
        else
        cout<<"yes";





    return 0;
}
```
