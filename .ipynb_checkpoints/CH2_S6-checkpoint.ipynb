```python
#include <iostream>
#include <iomanip>

using namespace std;

/* 2
+
3
5
console => 2, + => "+", 3 => 3.0 , 5 => 5.0
*/

int main()
{
string amalgar;
int A, k;
float a1, a2=0, result;
k=0;
cin>>A>>amalgar;
while (k < A){
    k+=1;
    cin>>a1;
    a2 += a1;
    // a1 = 3 , a2 = 3, a1 = 5 , a2 = 8, a2 = 3
}
a2 = a2 - a1;
if (amalgar=="+")
  result= a1+a2;
else if (amalgar=="ABS"){
    if (a1<0)
        result=-a1;
    else
        result=a1;
}
else if (amalgar=="INV")
    result=1.0/a1;
else
result=a1/a2;
cout<< "result : " << result;

    return 0;
}


```

```python
#include <iostream>
#include <iomanip>

using namespace std;


int main()
{
    int n, i, sum=1;
    cin>> n;
    for (i=2; i<n; i=i+1){
        if (n%i==0)
        sum=sum+i;
    }
    if (sum==n)
        cout<<"n is complete";
        else
         cout<<"n is not complete";
    return 0;
}


```

```python
#include <iostream>
#include <iomanip>

using namespace std;

// 563 => 3, 56 => 6, 5 => 5, 0


int main()
{
    int n, sum=0, i;
    cin>>n;
    while(n>0){
        sum=sum+n%10;
    n=n/10;
    }
   bool c=true;
   for (i=2; i<sum; i++){
        if (sum%i==0){
        c=false;
        break;
        }
   }
   if (c==true)
    cout<< "sum is prime";
    else
    cout << "sum is not prime";
    return 0;
}

```
