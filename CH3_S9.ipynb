```python
#include <iostream>
using namespace std;

int f(int n)
{
    if (n == 1 || n == 2)
        return 1;
    return f(n-1)+f(n-2);
}

// 1 , 1, f(2) + f(1)= 1+ 1 = 2

bool check_fib(int n)
{
    int counter;
    counter = 1;
    while (f(counter) < n)
    {
        counter += 1;
    }
    if (f(counter) == n)
     return true;
    else
        return false;

}


int main()
{
    int n, counter;
    cin >> n;
    while(n!=0){
        if (check_fib(n) == 1){
            cout << n;
            break;
        }
        else
            n = n - 1;
    }

    return 0;
}
```
